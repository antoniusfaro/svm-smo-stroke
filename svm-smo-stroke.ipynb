## Cross Validation

### SVM-SMO

```python
class Support_Vector_Machine:
    def __init__(self, C, gamma, tol = 0.001,  eps = 1e-05, verbose=False):
        self.C = C
        self.gamma = gamma
        self.tol = tol
        self.eps = eps
        self.iteration = 1
        self.verbose = verbose
    
    def latih(self, X, y):
        self.X = X
        self.y = y
        self.m = X.shape[0]
        self.n = X.shape[1]

        self.alpha, self.b = self.inisialisasi_alpha_b()       
        self.error_cache = self.inisialisasi_error_cache(self.X, self.y)
        
        cek_semua = 1
        jumlah_berubah = 0        
        
        while (jumlah_berubah > 0) or (cek_semua):            
            jumlah_berubah = 0
            if cek_semua:
                self.verbose and print('\nPeriksa Semua')
                for j in range(self.alpha.shape[0]):
                    self.verbose and print(f'\nj: {j}')
                    self.verbose and print('Iterasi ke: ', self.iteration)
                    hasil_opt = self.periksa_sampel(j)
                    jumlah_berubah = jumlah_berubah + hasil_opt
            
            else:
                for j in np.where((self.alpha != 0) & (self.alpha != self.C))[0]:                    
                    self.verbose and print(f'\nj: {j}')
                    self.verbose and print('\nIterasi ke: ', self.iteration)
                    hasil_opt = self.periksa_sampel(j)
                    jumlah_berubah = jumlah_berubah + hasil_opt
            
            if cek_semua == 1:
                cek_semua = 0
            elif jumlah_berubah == 0:
                cek_semua = 1
        self.verbose and print(f'alpha {self.alpha}')
        self.verbose and print(f'b {self.b}')
        print(f'total iterartion {self.iteration}')
                
    def periksa_sampel(self, j):
        yj = self.y[j]
        aj = self.alpha[j]
        ej = self.error_cache[j]

        if (((ej * yj) < -self.tol and aj < self.C) or ((ej * yj) > self.tol and aj > 0)):
            self.verbose and print(f'\n\tj: {j} MELANGGAR KKT')
            if len(self.alpha[(self.alpha != 0) & (self.alpha != self.C)]) > 1:
                if self.error_cache[j] > 0:
                    i = np.argmin(self.error_cache)
                    self.verbose and print(f'\n\tHEU-2 = MIN: {self.error_cache[i]}')
                else:
                    i = np.argmax(self.error_cache)
                    self.verbose and print(f'\n\tHEU-2 = MAX: {self.error_cache[i]}')
                hasil_opt = self.optimisasi(i, j)
                self.iteration += 1
                if hasil_opt:
                    return 1
        
            for i in np.roll(np.where((self.alpha != 0) & (self.alpha != self.C))[0], 
                             np.random.choice(np.arange(self.m))):
                self.verbose and print(f'\n\tHEU-2 = ROLL SV: {i}')
                hasil_opt = self.optimisasi(i, j)
                self.iteration += 1
                if hasil_opt:
                    return 1
        
            for i in np.roll(np.arange(self.m), np.random.choice(np.arange(self.m))):
                self.verbose and print(f'\n\tHEU-2 = ROLL ALL: {i}')
                hasil_opt = self.optimisasi(i, j)
                self.iteration += 1
                if hasil_opt:
                    return 1
        self.verbose and print(f'\n\tj: {j} MEMENUHI KKT')
        self.iteration += 1
        return 0
    
    def optimisasi(self, i, j):
        self.verbose and print(f'\nITERASI KE: {self.iteration}')
        self.verbose and print(f'\n\tTERPILIH: i: {i}, j: {j}')
        if i == j:
            self.verbose and print('\t\tGAGAL! i == j')
            return 0
        
        ai = self.alpha[i]
        aj = self.alpha[j]
        yi = self.y[i]
        yj = self.y[j]
        ei = self.error_cache[i]
        ej = self.error_cache[j]
        self.verbose and print(f'\t\tXi: {self.X[i]}')
        self.verbose and print(f'\t\tXj: {self.X[j]}')
        self.verbose and print(f'\t\tai OLD: {ai}')
        self.verbose and print(f'\t\taj OLD: {aj}')
        self.verbose and print(f'\t\tyi: {yi}')
        self.verbose and print(f'\t\tyj: {yj}')
        self.verbose and print(f'\t\tei: {ei}')
        self.verbose and print(f'\t\tej: {ej}')
        self.verbose and print(f'\t\tfxi: {ei + yi}')
        self.verbose and print(f'\t\tfxj: {ej + yj}')
        
        L, H = self.hitung_L_dan_H(yi, yj, ai, aj, self.C)
        self.verbose and print(f'\t\tL: {L} H: {H}')
        if L == H:
            self.verbose and print('\t\tGAGAL! L == H')
            return 0
        kii, kjj, kij, eta = self.hitung_eta(self.X[i], self.X[j])
        self.verbose and print(f'\n\t\tkii: {kii}, kjj: {kjj}, kij: {kij}, \neta: {eta}')
        aj_baru = self.hitung_aj_baru(aj, yj, ei, ej, eta)
        self.verbose and print(f'aj_baru: {aj_baru}')
        aj_baru = self.pangkas_aj_baru(aj_baru, L, H)
        self.verbose and print(f'aj_baru_pangkas: {aj_baru}')
                
        if np.abs(aj_baru - aj) < (self.eps * (aj_baru + aj + self.eps)):
            self.verbose and print('\n\t\tGAGAL! EPSILON!')
            return 0
        
        ai_baru = self.hitung_ai_baru(ai, aj, aj_baru, yi, yj)
        self.verbose and print(f'\n\t\tai_barunew: {ai_baru}')
        b1, b2 = self.hitung_b1_b2(ai, ai_baru, aj, aj_baru, yi, yj, ei, ej, kii, kjj, kij, self.b)        
        self.verbose and print(f'\n\t\tb1: {b1}, b2: {b2}')
        b_baru = self.hitung_b_baru(ai_baru, aj_baru, b1, b2, self.C)
        self.verbose and print(f'\t\tbnew: {b_baru}')
        self.alpha[i] = ai_baru
        self.alpha[j] = aj_baru
        
        self.hitung_error_cache(i, j, ai, ai_baru, aj, aj_baru, yi, yj, b_baru)
        self.b = b_baru
        self.verbose and print(f'\n\t\terror after ei:{self.error_cache[i]}, ej:{self.error_cache[j]}')
        self.verbose and print('\nERROR CACHE BARU:', i, ':', self.error_cache[i],
             ',', j, self.error_cache[j], 'last:', self.error_cache[-1])
        self.verbose and print(f'\n\tSUCCESS: i: {i}, j: {j}')
        return 1
    
    def inisialisasi_alpha_b(self):        
        alpha = np.zeros(self.m)
        b = 0.0
        return alpha, b
    
    def inisialisasi_error_cache(self, X, y):
        error_cache = self.fungsi_keputusan(X) - y
        return error_cache
    
    def hitung_L_dan_H(self, yi, yj, ai, aj, C):
        if (yi == yj):
            L = max(0, ai + aj - C)
            H = min(C, ai + aj)
        else:
            L = max(0, aj - ai)
            H = min(C, C + aj - ai)
        return L, H
    
    def hitung_eta(self, xi, xj):
        kii = self.hitung_kernel(xi, xi)
        kjj = self.hitung_kernel(xj, xj)
        kij = self.hitung_kernel(xi, xj)
        eta = 2 * kij - kii - kjj
        return kii, kjj, kij, eta
    
    def hitung_aj_baru(self, aj, yj, ei, ej, eta):
        aj_baru = aj - ((yj * (ei - ej)) / eta)
        return aj_baru
    
    def pangkas_aj_baru(self, aj_baru, L, H):
        if aj_baru <= L:
            aj_baru = L
        elif aj_baru >= H:
            aj_baru = H
        return aj_baru
    
    def hitung_ai_baru(self, ai, aj, aj_baru, yi, yj):
        ai_baru = ai + (yi * yj * (aj - aj_baru))
        return ai_baru
        
    def hitung_b1_b2(self, ai, ai_baru, aj, aj_baru, yi, yj, ei, ej, kii, kjj, kij, b):
        b1 = ei + (yi * (ai_baru - ai) * kii) + (yj * (aj_baru - aj) * kij) + b
        b2 = ej + (yi * (ai_baru - ai) * kij) + (yj * (aj_baru - aj) * kjj) + b
        return b1, b2
    
    def hitung_b_baru(self, ai_baru, aj_baru, b1, b2, C):
        if ai_baru > 0 and ai_baru < self.C:
            b_baru = b1
        elif aj_baru > 0 and aj_baru < self.C:
            b_baru = b2
        else:
            b_baru = 0.5 * (b1 + b2)            
        return b_baru    
    
    def hitung_error_cache(self, i, j, ai, ai_baru, aj, aj_baru, yi, yj, b_baru):
        for index, alph in zip([i, j], [ai_baru, aj_baru]):
            if 0.0 < alph < self.C:
                self.error_cache[index] = 0
        non_opt = [n for n in range(self.m) if (n != i and n!= j)]
        self.error_cache[non_opt] += \
            (yi * (ai_baru - ai) * self.hitung_kernel(self.X[i], self.X[non_opt])) + \
            (yj * (aj_baru - aj) * self.hitung_kernel(self.X[j],self.X[non_opt])) + \
            self.b - b_baru
    
    def fungsi_keputusan(self, X_test):
        hasil_fungsi_keputusan = (self.alpha * self.y) @ self.hitung_kernel(self.X, X_test) - self.b
        return hasil_fungsi_keputusan
        
    def hitung_kernel(self, x, y):
        if np.ndim(x) == 1 and np.ndim(y) == 1:
            return np.exp(- self.gamma * (np.linalg.norm(x - y, 2)) ** 2)
        elif (np.ndim(x) == 1 and np.ndim(y) > 1) or (np.ndim(x) > 1 and np.ndim(y) == 1):
            return np.exp(- self.gamma * (np.linalg.norm(x - y, 2, axis = 1)) ** 2)
        else:
            return np.exp(- self.gamma * (np.linalg.norm(x[:, np.newaxis] - y[np.newaxis, :], 2, axis = 2)) ** 2)
    
    def prediksi(self, X):
        hasil_fungsi_keputusan = self.fungsi_keputusan(X)
#         print(hasil_fungsi_keputusan)
        hasil_prediksi = np.where(hasil_fungsi_keputusan < 0, -1, 1)
        return hasil_prediksi
    
    @staticmethod
    def metrics_calc(y, y_pred):
        accuracy = np.round((np.sum(y == y_pred) / y.shape[0]) * 100, 2)
        return accuracy
    
    def __del__(self):
        print('')
```

```python
np.random.seed(0)
svm = Support_Vector_Machine(C = 10, gamma=0.0005, verbose=False)
# start = time.time()        
svm.latih(X_train, y_train)
# end = time.time()

y_train_pred = svm.prediksi(X_train)
y_test_pred = svm.prediksi(X_test)
print(y_test)
print(y_test_pred)
train_acc = Support_Vector_Machine.metrics_calc(y_train, y_train_pred)
test_acc = Support_Vector_Machine.metrics_calc(y_test, y_test_pred)

# print('Time: ', ((end-start) // 60), 'min', ((end-start) % 60), 'sec')
print('Train accuracy with C: {} gamma: {} is {}'.format(10, 0.0005, train_acc))
print('Test accuracy with C: {} gamma: {} is {}'.format(10, 0.0005, test_acc))
```
